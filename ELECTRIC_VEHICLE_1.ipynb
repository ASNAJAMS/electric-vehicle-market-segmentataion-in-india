```python




# Create categories for Charging Stations
data_ev['Charging_Station_Category'] = pd.cut(
    data_ev['Charging Stations'],
    bins=[0, 100, 300, 1000, float('inf')],
    labels=['Low', 'Medium', 'High', 'Very High']
)

# Create categories for EV Adoption (Proportion of Electric Vehicles)
data_ev['EV_Adoption_Category'] = pd.cut(
    data_ev['Proportion of Electric Vehicles (%)'],
    bins=[0, 0.5, 1.0, 2.0, float('inf')],
    labels=['Low', 'Medium', 'High', 'Very High']
)

# Create categories for FinTech Hub presence
data_ev['FinTech_Category'] = pd.cut(
    data_ev['Number of FinTechs'],
    bins=[0, 10, 50, 100, float('inf')],
    labels=['Low', 'Medium', 'High', 'Very High']
)

# Define a function to assign urban, semi-urban, or rural based on charging stations and FinTech presence
def assign_geographical_segment(row):
    if row['Charging_Station_Category'] in ['High', 'Very High'] or row['FinTech_Category'] in ['High', 'Very High']:
        return 'Urban'
    elif row['Charging_Station_Category'] == 'Medium' or row['FinTech_Category'] == 'Medium':
        return 'Semi-Urban'
    else:
        return 'Rural'

# Apply the function to create a new column for Geographical Segment
data_ev['Geographical_Segment'] = data_ev.apply(assign_geographical_segment, axis=1)

# Display the dataset with new segments
data_ev[['State/UT', 'Charging Stations', 'Proportion of Electric Vehicles (%)', 'Number of FinTechs', 'Charging_Station_Category', 'EV_Adoption_Category', 'FinTech_Category', 'Geographical_Segment']]

# You can further analyze or plot these segments based on the new 'Geographical_Segment' category
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming the dataset 'data_ev' contains relevant columns for geographical segmentation

# Plotting the Geographical Segments for Charging Stations, EV Adoption, and FinTechs
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Charging Stations', y='Total EVs Registered', hue='Geographical_Segment', 
                style='EV_Adoption_Category', size='Number of FinTechs', data=data_ev, 
                palette='Set1', sizes=(40, 200))
plt.title('Geographical Segmentation based on Charging Stations, EV Adoption, and FinTech Presence')
plt.xlabel('Charging Stations')
plt.ylabel('Total EVs Registered')
plt.legend(title='Geographical Segment')
plt.show()
```

```python

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans


# Assume your dataset is already loaded into 'data_ev'

# Step 1: Encode the categorical columns
label_encoders = {}
for col in ['Charging_Station_Category', 'EV_Adoption_Category', 'FinTech_Category']:
    le = LabelEncoder()
    data_ev[col + '_Encoded'] = le.fit_transform(data_ev[col].astype(str))  # Encode the categories
    label_encoders[col] = le

# Step 2: Select relevant columns including the encoded categorical columns
X = data_ev[['Charging Stations', 'Proportion of Electric Vehicles (%)', 'Number of FinTechs',
             'Charging_Station_Category_Encoded', 'EV_Adoption_Category_Encoded', 'FinTech_Category_Encoded']]

# Step 3: Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Apply PCA for dimensionality reduction
pca = PCA(n_components=2)  # Reducing to 2 principal components
X_pca = pca.fit_transform(X_scaled)

# Step 5: Apply KMeans for clustering based on the PCA results
kmeans = KMeans(n_clusters=4, random_state=42)
data_ev['Cluster'] = kmeans.fit_predict(X_pca)

# Step 6: Visualize the results of PCA with clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=data_ev['Cluster'], palette='Set1', s=100)
plt.title('Geographical Segmentation using PCA and KMeans Clustering (with Encoded Categories)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()

# Step 7: Display the final DataFrame with clusters
print(data_ev[['State/UT', 'Cluster']])
```

```python


# Define colors for the labels based on cluster
label_colors = {0: 'red', 1: 'blue', 2: 'green', 3: 'purple'}

# Increase marker size and add labels with colored text
plt.figure(figsize=(12, 8))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=data_ev['Cluster'], palette='Set1', s=150, edgecolor='k')

# Add labels for each point based on the State/UT column, with colors based on clusters
for i, txt in enumerate(data_ev['State/UT']):
    plt.text(X_pca[i, 0] + 0.02, X_pca[i, 1], txt, fontsize=9, weight='bold', color=label_colors[data_ev['Cluster'][i]],
             bbox=dict(facecolor='white', alpha=0.6, edgecolor='none'))

# Add titles and labels
plt.title('Geographical Segmentation with PCA and KMeans Clustering', fontsize=14)
plt.xlabel('Principal Component 1', fontsize=12)
plt.ylabel('Principal Component 2', fontsize=12)
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Assuming 'data_ev' contains the following columns: 'E-2W Incentive (INR/kWh)', 'E-3W Incentive (INR/kWh)', 'E-4W Incentive (INR/kWh)'

# First, we need to convert the percentage values (e.g., "15%") in the incentive columns to numeric values
def convert_to_numeric(value):
    if isinstance(value, str) and '%' in value:
        return float(value.replace('%', '')) / 100
    return float(value) if pd.notnull(value) else 0

# Apply the conversion function to all relevant incentive columns
data_ev['E-2W Incentive (Numeric)'] = data_ev['E-2W Incentive (INR/kWh)'].apply(convert_to_numeric)
data_ev['E-3W Incentive (Numeric)'] = data_ev['E-3W Incentive (INR/kWh)'].apply(convert_to_numeric)
data_ev['E-4W Incentive (Numeric)'] = data_ev['E-4W Incentive (INR/kWh)'].apply(convert_to_numeric)

# Now, aggregate the incentive columns into a single 'Government Incentives' score (summing them up)
data_ev['Government Incentives'] = (
    data_ev['E-2W Incentive (Numeric)'] + 
    data_ev['E-3W Incentive (Numeric)'] + 
    data_ev['E-4W Incentive (Numeric)']
)

# Display the new Government Incentives column
print(data_ev[['State/UT', 'Government Incentives']].head(15))
```
